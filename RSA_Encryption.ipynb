### आर् एस् ए इन्क्रिप्सन 

```python
import pyinputplus as pyip
import math, random

# checking input is prime or not
def is_prime(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    if count == 2:
        return True
    else:
        return False
#     calculating d for the private key
def calculate_d(totient_n, e):
    for i in range(2, totient_n):
        if (e * i) % totient_n == 1:
            return i
    
#     enter two inputs to be prime
while True:
    print("Enter two prime numbers(रूढ संख्याहरू ): ")
    p = pyip.inputNum()
    q = pyip.inputNum()
    if is_prime(p) and is_prime(q) and p != q:
        break

# calculate n
n  = p * q
print("n: ", n)
# calculate Euler Totient function
totient_n = (p - 1) * (q - 1)
print("totient(n): ", totient_n)
e_list = []
for i in range(2, totient_n):
    if math.gcd(totient_n, i) == 1:
        e_list.append(i)
# print("e_list: ", e_list)
while True:
#     e = pyip.inputInt("Select an integer from e_list: ")
    e = random.choice(e_list)
    if e in e_list:
        break
print("e: ", e)

# calculate d
d = calculate_d(totient_n, e)
print("d: ", d)
# public key
PU = [e, n]
print("Public key: ", PU)
# private key
PR = [d, n]
print("Private key: ", PR)
# plaintext
m = pyip.inputInt("Enter message: ", lessThan = n)
# Encryption
c = m ** e % n
# c = math.fmod(pow(m, e), n)
print("Cipher text: ", c)
# Decryption
m = c ** d % n
print("Decrypted text: ", m)
```

## RSA Encryption With Tkinter

```python
from tkinter import *
import pyinputplus as pyip
import math, random

# checking input is prime or not
def is_prime(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    if count == 2:
        return True
    else:
        return False
#     calculating d for the private key
def calculate_d(totient_n, e):
    for i in range(2, totient_n):
        if (e * i) % totient_n == 1:
            return i

class RSA(Frame):
    def __init__(self, master):
        super(RSA, self).__init__(master)
        self.grid()
        self.create_widget()
    
    def create_widget(self):
        Label(self,
              text = 'Enter two prime numbers(रूढ संख्याहरू): '
             ).grid(row = 0, column = 0, sticky = W)
        
        self.p_entry = Entry(self, width = 5)
        self.p_entry.grid(row = 0, column = 1)
        
        self.q_entry = Entry(self, width = 5)
        self.q_entry.grid(row = 0, column = 3, padx = 10)
        
        self.submit_primes = Button(self, text = 'Submit', command = self.display_calc)
        self.submit_primes.grid(row = 1, column = 0)
        
        self.reset_primes = Button(self, text = 'Reset', command = self.remove_error_text)
        self.reset_primes.grid(row = 1, column = 1)
    
    def display_calc(self):
        p = int(self.p_entry.get())
        q = int(self.q_entry.get())
        
        if not is_prime(p) or not is_prime(q) or p == q:
            self.not_primes_error = Label(self, text = 'Both entry must be prime.')
            self.not_primes_error.grid(row = 2, column = 0, sticky = W)

            self.equal_error = Label(self, text = 'Entries should not be equal.')
            self.equal_error.grid(row = 3, column = 0, sticky = W)
        
        else:
            self.n = p * q
            Label(self,
                  text = f'n: {self.n}'
                 ).grid(row = 2, column = 0, sticky = W)
            
            totient_n = (p - 1) * (q - 1)
            Label(self,
                  text = f'totient(n): {totient_n}'
                 ).grid(row = 3, column = 0, sticky = W)
            
            e_list = []
            for i in range(2, totient_n):
                if math.gcd(totient_n, i) == 1:
                    e_list.append(i)
            
            self.e = random.choice(e_list)
            Label(self, 
                  text = f'e: {self.e}'
                 ).grid(row = 4, column = 0, sticky = W)
    
            # calculate d
            self.d = calculate_d(totient_n, self.e)
            Label(self,
                  text = f'd: {self.d}'
                 ).grid(row = 5, column = 0, sticky = W)

            # public key
            PU = [self.e, self.n]
            Label(self,
                  text = f'Public Key: {PU}'
                 ).grid(row = 6, column = 0, sticky = W)

            # private key
            PR = [self.d, self.n]
            Label(self,
                  text = f'Private Key: {PR}'
                 ).grid(row = 7, column = 0, sticky = W)

            # plaintext
            Label(self,
                  text = 'Enter message(in integer and less than n): '
                 ).grid(row = 8, column = 0, sticky = W)
            self.message = Entry(self, width = 5)
            self.message.grid(row = 8, column = 1)
            
            # submit message
            self.submit_msg = Button(self, text = 'Submit Msg', command = self.encrypt_msg)
            self.submit_msg.grid(row = 9, column = 0)
            
    def remove_error_text(self):
        self.p_entry.delete(0, END)
        self.p_entry.insert(0, '')
        self.q_entry.delete(0, END)
        self.q_entry.insert(0, '')
        self.equal_error['text'] = ''
        self.not_primes_error['text'] = ''
    
    def encrypt_msg(self):
        msg = int(self.message.get())
        # Encryption
        c = msg ** self.e % self.n
        Label(self, 
              text = f'Cipher text: {c}'
             ).grid(row = 10, column = 0, sticky = W)

        # Decryption
        m = c ** self.d % self.n
        Label(self,
              text = f'Decrypted text: {m}'
             ).grid(row = 11, column = 0, sticky = W)         

        
root = Tk()
root.title('RSA Encryption')
root.geometry('500x500+50+50')
app = RSA(root)
root.mainloop()
```
